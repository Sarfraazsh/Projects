```python
claims_df = pd.read_csv("/Users/sarfraazshahulhameed/Documents/Assignments/HAP 720/hap720_claims_dgn_demo_distorted.csv", low_memory = False)
seer_df = pd.read_csv("/Users/sarfraazshahulhameed/Documents/Assignments/HAP 720/hap720_seer_distorted.csv", low_memory = False)
```

```python
# Convert the columns in the seer DataFrame to object type
seer_df['diagnosis_year'] = seer_df['diagnosis_year'].astype('object')
seer_df['diagnosis_age'] = seer_df['diagnosis_age'].astype('object')
seer_df['c_diagnosis'] = seer_df['c_diagnosis'].astype('object')
```

```python
# Merge the two DataFrames based on matching columns
merged_df = pd.merge(claims_df, seer_df,
                     left_on=['claim_year', 'age', 'sex', 'race', 'diagnosis9'],
                     right_on=['diagnosis_year', 'diagnosis_age', 'sex', 'race', 'c_diagnosis'],
                     how='outer')
```

```python
# Select the desired columns and rename them
merged_df = merged_df[['patient_id', 'claim_year', 'claim_date', 'death_date', 'sex',
                       'age', 'race', 'diagnosis9', 'diagnosis10', 'claim_number', 'member_id']]
merged_df.columns = ['patient_id', 'claim_year', 'claim_date', 'death_date', 'sex',
                     'age', 'race', 'diagnosis9', 'diagnosis10', 'claim_number', 'member_id']
```
